```python


# Read the concatenated CSV file into a DataFrame
df = pd.read_csv('concatenated.csv')

# Display information about the DataFrame
print("DataFrame Information:")
print(df.info())
print("\n")

# Generate descriptive statistics of the DataFrame
print("DataFrame Statistics:")
print(df.describe().T)
print("\n")

```

```python
# Visualize missing values using missingno library
msno.heatmap(df)

msno.matrix(df)
```

```python
df = df[['Group','Age','M/F','EDUC','SES','MMSE','CDR','eTIV','nWBV','ASF']]
```

```python
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
```

```python
df['M/F'] = df['M/F'].replace(['F','M'], [0,1])
df['Group'] = df['Group'].replace(['Converted'], ['Demented'])
df['Group'] = df['Group'].replace(['Demented', 'Nondemented'], [1,0])
```

```python
# Import the necessary library for advanced imputation
from sklearn.impute import KNNImputer

# Create an instance of the KNNImputer
imputer = KNNImputer()

# Apply the imputation to fill missing values
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
# columns_to_convert = ['Group','Age','M/F','eTIV']

# for column in columns_to_convert:
#     df_imputed[column] = df_imputed[column].astype(int)
    
# df_imputed.to_csv('concatenated_imputed.csv', index=False)
```

```python
grouped_data = df_imputed.groupby(['Age', 'M/F'])['CDR'].mean().unstack()

# Set the plot size
# Plot the grouped data as a bar plot
grouped_data.plot(kind='bar',width=0.5,figsize=(20,10))
# Set plot labels and title
plt.xlabel('Age')
plt.ylabel('CDR')
plt.title('CDR by Age and Gender')

# Rotate x-axis tick labels if necessary
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming your imputed data is stored in a DataFrame called 'df_imputed'
columns = df_imputed.columns  # Get the column names

num_cols = len(columns)  # Number of columns in your data

# Calculate the number of rows and columns for the grid
num_rows = (num_cols + 2) // 3  # Adjust the number based on the desired layout
num_cols = min(num_cols, 3)

# Create a figure and subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 12))

# Flatten the axs array if it has more than one dimension
if num_rows > 1:
    axs = axs.flatten()

# Iterate over the columns and create plots
for i, col in enumerate(columns):
    ax = axs[i]  # Select the appropriate subplot
    ax.hist(df_imputed[col])  # Create the box plot for the current column
    ax.set_xlabel('')  # Set the x-axis label
    ax.set_ylabel('count')  # Set the y-axis label
    ax.set_title(col)  # Set the title of the subplot

# Hide any extra subplots
for j in range(i+1, len(axs)):
    axs[j].axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
